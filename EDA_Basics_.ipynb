```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/House Price India DATA ASSIGNMENT-1.csv")
df.head()
#df.info()
#df.describe()
#df.isnull().sum().sum()
```

```python
# Q1. What is the average, median, and standard deviation of house prices?
avg_price = df['Price'].mean()
median_price = df['Price'].median()
std_price = df['Price'].std()
print("avarage price :", avg_price)
print("median price :", median_price)
print("standard deviation price :", std_price)

```

**Interpretation:** The average provides the center of the data but can be pulled by extremes. The median gives the exact middle point of the market, while the standard deviation tells us how much the house prices typically deviate from the average (measuring market volatility/spread).

```python
# Q2. Which number of bedrooms is most common?
most_common_bedrooms = df['number of bedrooms'].mode()[0]

print("Most common number of bedrooms:", most_common_bedrooms)


```

**Interpretation:** Identifying the mode helps real estate developers understand consumer demand trends. The most common bedroom count represents the "typical" house configuration in this market.

```python
# Q3. Check if the price data is skewed.
Price_skewness = df['Price'].skew()
print("Price Skewness:", Price_skewness)
if Price_skewness > 0:
  print("The Price data is positively skewed.")
elif Price_skewness < 0:
    print("The Price data is Negatively skewed.")
else:
  print("The Price data is Perfectly symmetrical. ")
```

**Interpretation:** Real estate prices are almost always positively skewed. This means the majority of houses are clustered around lower to middle price points, with a few highly expensive luxury properties pulling the tail to the right.

```python
# Q4. What is the average price for each number of bedrooms?
avg_price_per_bedrooms = df.groupby('number of bedrooms')['Price'].mean().reset_index()
print("Average price for each number of bedrooms:")
print(avg_price_per_bedrooms)
```

**Interpretation:** This highlights the marginal valuation increase per additional bedroom. Typically, prices scale upward with bedroom count, though extreme counts (e.g., 0 or 10+ bedrooms) might display anomalous behavior due to unique property types.

```python
# Q5. What is the relationship between living area and price? (Use: Area without Basement).
correlation = df['living area'].corr(df['Price'])
print(f"correlation between living area and price:", correlation)
```

**Interpretation:** A positive correlation close to $1$ implies that as the ground/above-ground living area increases, the house price increases significantly. This is usually one of the strongest predictors of real estate value.

```python
# Q6. Identify any anomalies where houses have high prices but low areas.
high_price_threshold = df['Price'].quantile(0.95)
low_area_threshold = df['living area'].quantile(0.05)
anomalies = df[(df['Price'] > high_price_threshold) & (df['living area'] < low_area_threshold)]
print(f"number of anomalies found:", len(anomalies))
print(anomalies[['Price', 'living area']].head())
```

**Interpretation:** These anomalies point to properties that buck the square-footage trend. This could represent ultra-luxury micro-apartments, historic buildings, or properties in highly premium, exclusive zip codes where location value completely overrides size.

```python
# Q7. Compare average price based on number of floors and houses with or without waterfront.
# Grouping by both floors and waterfront status
pivot_table = df.groupby(['number of floors', 'waterfront present'])['Price'].mean().unstack()
print(round(pivot_table, 2))
```

**Interpretation:** This breakdown lets us see if premium features like a waterfront view multiply value evenly across all property types (single-story vs. multi-story) or if it benefits premium multi-floored assets more heavily.

```python
# Q8. Identify the minimum and maximum house price. What does this indicate?
min_price = df['Price'].min()
max_price = df['Price'].max()
print(f"minimum house price :", min_price)
print(f"maximum house price :", max_price)
print(f"price range :", {max_price - min_price})
```

**Interpretation:** A massive gap between the minimum and maximum prices indicates a highly diverse real estate market, spanning low-income affordable housing up to ultra-luxury estates.

```python
# Q9. Which location (zipcode/area) has the highest average price?
highest_price_per_location = df.groupby('Postal Code')['Price'].mean().idxmax()
highest_avg_price =  df.groupby('Postal Code')['Price'].mean().max()
print("Location with the highest average price:", highest_price_per_location)
print("Highest average price:", highest_avg_price)
```

**Interpretation:** This identifies the most premium/posh neighborhood or pin code in the dataset, which is critical geographic insight for real estate investment decisions.
