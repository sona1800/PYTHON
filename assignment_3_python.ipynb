# Q4. Indexing using .loc

```python
# row with index 10
print(df.loc[10])

# sales value of the row with index 20
print("Sales at index 20:", df.loc[20, "Sales"])
```

# Q9. Descriptive Statistics

Generate summary statistics using:

describe()

From the result identify:

Mean sales value

Maximum quantity sold

Minimum sales value



```python
stats = df.describe()
print(stats)

# result identify
print("mean Sales:", stats.loc['mean', 'Sales'])
print("max Sales:", stats.loc['max', 'Sales'])
print("min Sales:", stats.loc['min', 'Sales'])
```

# Q10. Aggregation Analysis

Using the Sales column, calculate:

Total sales

Average sales

Maximum sales

Minimum sales

```python
total_sales = df['Sales'].sum()
print("Total Sales:", total_sales)

avg_sales = df['Sales'].mean()
print("Average Sales:", avg_sales)

max_sales = df['Sales'].max()
print("Maximum Sales:", max_sales)

min_sales = df['Sales'].min()
print("Minimum Sales:", min_sales)

```
