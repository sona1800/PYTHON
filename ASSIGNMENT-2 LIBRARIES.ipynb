```python
import pandas as pd
import numpy as np
```

#### Q.1

```python
df = pd.read_csv(r"C:\data analyst\python\retail_dataset assignment 2.csv")
df.head()
df.info()	
```

#### Q2: Identify missing values

```python
# Q2: Identify missing values
missing_per_col = df.isnull().sum()
total_missing = df.isnull().sum().sum()
print(f"Missing values per column:\n{missing_per_col}")
print(f"Total missing values: {total_missing}")

```

```python
# Q3: Replace missing Quantity and Price with mean
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].mean())
df['Price'] = df['Price'].fillna(df['Price'].mean())


```

```python
# Q4: Remove rows with missing Product Category or Region
df.dropna(subset=['Product Category', 'Region'], inplace=True)
```

```python
# Q5: Create Revenue column using NumPy
# Revenue = Quantity * Price
df['Revenue'] = np.multiply(df['Quantity'], df['Price'])
print(df.head())

```

```python
# Q6: Calculate total revenue
total_revenue = np.sum(df['Revenue'])
print(f"Total Revenue: {total_revenue}")
```

```python
# Q7: Revenue by Product Category
cat_revenue = df.groupby('Product Category')['Revenue'].sum()
print(cat_revenue)

```

```python
# Q8: Top 3 and Bottom 3 Categories
top_3 = cat_revenue.nlargest(3)
bottom_3 = cat_revenue.nsmallest(3)
print(f"Top 3:\n{top_3}\nBottom 3:\n{bottom_3}")

```

```python
# Q9: Revenue by Region
region_revenue = df.groupby('Region')['Revenue'].sum()
highest_region = region_revenue.idxmax()
lowest_region = region_revenue.idxmin()
print(f"Highest Region: {highest_region}, Lowest Region: {lowest_region}")
```

```python
# Q10. Monthly analysis and NumPy statistics
# Extract month from 'Date'
import pandas as pd
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

# Group by month for total revenue
monthly_revenue = df.groupby('Month')['Revenue'].sum()
print("\n--- Q10: Total Monthly Revenue ---")
print(monthly_revenue)

# Calculate Mean, Median, and Standard Deviation using NumPy
mean_val = np.mean(df['Revenue'])
median_val = np.median(df['Revenue'])
std_dev_val = np.std(df['Revenue'])

print(f"\n--- Q10: Statistics for Revenue Column ---")
print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"Standard Deviation: {std_dev_val:.2f}")
```
