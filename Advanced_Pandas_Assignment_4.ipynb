# Q2 Sales Filtering Analysis

Filter the dataset to display:

Orders where Sales > 3500

Orders where Region = "East" and Sales > 2500

Explain what insights you observe.

```python
#Orders where Sales > 3500
high_sales = df[df['Sales'] > 3500]
print(high_sales)

# Orders where Region = "East" and Sales > 2500
east_high_sales = df[(df['Region'] == 'East') & (df['Sales'] > 2500)]

print(east_high_sales)
```

```python
# Q3 Customer Purchase Analysis

# Identify the top 5 customers based on total sales.

top_5_customers = df.groupby('Customer')['Sales'].sum().sort_values(ascending=False).head(5)

print(top_5_customers)
```

```python
# Q4 Regional Sales Performance

# Calculate the total sales generated by each region.

#Then identify: Which region generated the highest revenue?




regional_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
highest_revenue_region = regional_sales.idxmax()

print(regional_sales)
print(f"The region with the highest revenue is: {highest_revenue_region}")

```

```python
# Q5 Product Performance Analysis

# Calculate the total quantity sold for each product.

# Which product appears to be the most popular?

product_quantity = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
most_popular_product = product_quantity.idxmax()

print(product_quantity)
print(f"The most popular product is: {most_popular_product}")

```

```python
# Q6 Multi-Level Grouping
# Group the dataset by: Region and Category
# Calculate the total sales for each combination.
# What insights can you derive from this analysis?

region_category_sales = df.groupby(['Region', 'Category'])['Sales'].sum()
print(region_category_sales)
```

Insight: This reveals which product categories drive revenue in specific geographic areas, helping with targeted inventory management.

```python
# Q7 Pivot Table Creation
#Create a pivot table showing:
#Rows → Region
#Columns → Product
#Values → Total Sales
#Which region sells the most laptops?
pivot_table = df.pivot_table(index='Region', columns='Product', values='Sales', aggfunc='sum')
print(pivot_table)
most_laptops_region = pivot_table['Laptop'].idxmax()
print(f"The region that sells the most laptops is: {most_laptops_region}")
```

```python
# Q8 Profitability Analysis
# Calculate the average profit for each product category.
# Which category is most profitable?

avg_profit = df.groupby('Category')['Profit'].mean().sort_values(ascending=False)
print(avg_profit)
print("\nMost Profitable Category:", avg_profit.idxmax())

```

```python
#Q9 Discount Impact Study
#Analyze the relationship between discount and profit.
#Tasks: Calculate the average profit for each discount level
#Identify whether higher discounts lead to lower profit

discount_profit = df.groupby('Discount')['Profit'].mean().sort_values(ascending=False)
print(discount_profit)

correlation = df['Discount'].corr(df['Profit'])
print(f"\nCorrelation between Discount and Profit: {correlation:.4f}")
```

Insights: There is a slight negative correlation (-0.0538). While lower discounts (0.0 or 0.1) show higher average profits, aggressive discounting (0.2+) starts to erode the profit margins significantly.


```python
#Q10 End-to-End Data Wrangling Task
#Perform the following workflow:
#Filter the dataset where Sales > 3000
#Group data by Region and Product
#Calculate total sales
#Reshape the result into a pivot table
#Explain the business insights obtained from this analysis.

# 1. Filter Sales > 3000
wrangled_df = df[df['Sales'] > 3000]

# 2. & 3. Group by Region/Product and Calculate total sales
grouped_wrangled = wrangled_df.groupby(['Region', 'Product'])['Sales'].sum().reset_index()

# 4. Reshape into pivot table
final_pivot = grouped_wrangled.pivot(index='Region', columns='Product', values='Sales').fillna(0)
print(final_pivot)
```
