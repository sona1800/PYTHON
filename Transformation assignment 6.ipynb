```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
df = pd.read_csv('/content/employee_productivity_dataset.csv')
df.head()
```

```python
#Q1. Handle Missing Values (Basic Cleaning)
#Fill missing values using the following:
#Age → Median
df_cleaned = df.copy()
df_cleaned['Age'] = df_cleaned['Age'].fillna(df_cleaned['Age'].median())
#Salary → Mean
df_cleaned['Salary'] = df_cleaned['Salary'].fillna(df_cleaned['Salary'].mean())
#Hours_Worked_Per_Week → Median
df_cleaned['Hours_Worked_Per_Week'] = df_cleaned['Hours_Worked_Per_Week'].fillna(df_cleaned['Hours_Worked_Per_Week'].median())
#Performance_Score → Mean
df_cleaned['Performance_Score'] = df_cleaned['Performance_Score'].fillna(df_cleaned['Performance_Score'].mean())
#Display the dataset after handling missing values.
print("Dataset after handling missing values:")
print(df_cleaned)
```

```python
# label Encoding
df_encoded = df_cleaned.copy()
le_gender = LabelEncoder()
le_dept = LabelEncoder()

df_encoded['Gender'] = le_gender.fit_transform(df_encoded['Gender'])
df_encoded['Department'] = le_dept.fit_transform(df_encoded['Department'])

print("\nQ2 Output - Updated Columns (Gender & Department):")
print(df_encoded[['Gender', 'Department']])
```

```python
# one hot encoding

df_ohe = pd.get_dummies(df_encoded, columns=['Work_Mode', 'Location'], dtype=int)

print("\nQ3 Output - Dataset with One-Hot Encoding:")
print(df_ohe)
print(f"\nNew columns created: {df_ohe.shape[1] - df_encoded.shape[1]}")
```

```python
# Normalization (min-max scaling)
df_normalized = df_ohe.copy()
min_max_scaler = MinMaxScaler()

columns_to_normalize = ['Salary', 'Hours_Worked_Per_Week']
df_normalized[columns_to_normalize] = min_max_scaler.fit_transform(df_normalized[columns_to_normalize])

print("\nQ4 Output - Normalized Salary and Hours Worked:")
print(df_normalized[columns_to_normalize])
```

```python
# Standardization (Z- score scaling)
df_standardized = df_normalized.copy()
std_scaler = StandardScaler()

columns_to_standardize = ['Age', 'Projects_Completed']
df_standardized[columns_to_standardize] = std_scaler.fit_transform(df_standardized[columns_to_standardize])

print("\nQ5 Output - Standardized Age and Projects Completed:")
print(df_standardized[columns_to_standardize])
```

Q9. Why is scaling important in Python?
Answer: Machine Learning algorithms calculate distances between data points (like KNN or SVM) or utilize gradient descent optimization (like Linear Regression and Neural Networks). If one feature has a massive range (e.g., Salary: $40,000–$100,000) and another has a small range (e.g., Age: 20–60), the larger variable will completely dominate the calculations. Scaling ensures every single feature contributes proportionately to the model’s learning process.

Q10. Why do we convert categorical data into numerical form?
Answer: Computer processors and mathematical equations cannot inherently calculate equations using text strings like "Remote" or "Sales". All underlying machine learning models are fundamentally mathematical algorithms that require matrix multiplication, weights, and numerical inputs to train and produce predictions.
